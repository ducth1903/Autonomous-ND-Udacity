```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

%matplotlib inline

image = mpimg.imread('bridge_shadow.jpg')
plt.imshow(image)
```

```python
# Gradient threshold 
def abs_sobel_thresh(img, orient='x', sobel_kernel=3, thresh=(0, 255)):   
    # Apply the following steps to img
    # 1) Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    
    # 2) Take the derivative in x or y given orient = 'x' or 'y'
    if orient=='x':
        grad = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=sobel_kernel)
    else:
        grad = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=sobel_kernel)
    
    # 3) Take the absolute value of the derivative or gradient
    grad = np.absolute(grad)
    
    # 4) Scale to 8-bit (0 - 255) then convert to type = np.uint8
    scaled = (grad*255/np.max(grad)).astype(np.uint8)
    
    # 5) Create a mask of 1's where the scaled gradient magnitude 
        # is > thresh_min and < thresh_max
    mask = np.zeros_like(scaled)
    mask[(scaled>=thresh[0]) & (scaled<=thresh[1])] = 1
    
    # 6) Return this mask as your binary_output image
    # binary_output = np.copy(img) # Remove this line
    binary_output = mask
   
    return binary_output
```

```python
# Color threshold
def hls_thresh(img, thresh_S):
    hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
    H, L, S = hls[:,:,0], hls[:,:,1], hls[:,:,2]
    binary_S = np.zeros_like(S)
    binary_S[(S>thresh_S[0]) & (S<=thresh_S[1])] = 1
    binary_output = binary_S
    return binary_output
```

```python
def pipeline(img, s_chan_thresh=(170,255), sobel_x_thresh=(20,100), isColor=False):
    img = np.copy(img)
    grad_binary = abs_sobel_thresh(img, orient='x', thresh=sobel_x_thresh)
    color_binary = hls_thresh(img, thresh_S=s_chan_thresh)
    
    if isColor:
        pipeline_img = np.dstack((np.zeros_like(color_binary), grad_binary, color_binary)) * 255
    else:
        pipeline_img = np.zeros_like(color_binary)
        pipeline_img[(grad_binary==1) | (color_binary==1)] = 1
    
    return pipeline_img

color_output = pipeline(image, isColor=True)
output = pipeline(image)

f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(24, 9))
f.tight_layout()
ax1.imshow(image, cmap='gray')
ax1.set_title('Original Image', fontsize=30)
ax2.imshow(color_output, cmap='gray')
ax2.set_title('Pipeline Color', fontsize=30)
ax3.imshow(output, cmap='gray')
ax3.set_title('Pipeline Black/white', fontsize=30)                 
plt.subplots_adjust(left=0., right=1, top=0.9, bottom=0.)
```
